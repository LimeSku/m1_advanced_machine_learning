With MNIST dataset

```python
# Load and preprocess the MNIST dataset
mnist = tf.keras.datasets.mnist
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = mnist.load_data()

# Reshape the data to include the channel dimension
x_train_mnist = x_train_mnist.reshape(-1, 28, 28, 1).astype('float32') / 255
x_test_mnist = x_test_mnist.reshape(-1, 28, 28, 1).astype('float32') / 255

# Define kernel and pool sizes
kernel_size = (3, 3)
pool_size = (2, 2)

batch_size = 32
num_classes = 10
epochs = 10

# Build the CNN model
model_mnist = models.Sequential()

# First convolutional layer
model_mnist.add(layers.Conv2D(32, kernel_size, padding='same', input_shape=x_train_mnist.shape[1:], activation='relu'))
model_mnist.add(layers.BatchNormalization())
model_mnist.add(layers.Conv2D(32, kernel_size, activation='relu'))
model_mnist.add(layers.BatchNormalization())
model_mnist.add(layers.MaxPooling2D(pool_size=pool_size))
model_mnist.add(layers.Dropout(0.25))

# Second convolutional layer
model_mnist.add(layers.Conv2D(64, kernel_size, padding='same', activation='relu'))
model_mnist.add(layers.BatchNormalization())
model_mnist.add(layers.Conv2D(64, kernel_size, activation='relu'))
model_mnist.add(layers.BatchNormalization())
model_mnist.add(layers.MaxPooling2D(pool_size=pool_size))
model_mnist.add(layers.Dropout(0.25))

# Flatten the output and add a fully connected layer
model_mnist.add(layers.Flatten())
model_mnist.add(layers.Dense(512, activation='relu'))
model_mnist.add(layers.BatchNormalization())
model_mnist.add(layers.Dropout(0.5))

# Output layer with softmax activation
model_mnist.add(layers.Dense(num_classes, activation='softmax'))

# Compile the model
model_mnist.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0001, decay=1e-06),
                    loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model_mnist.summary()

# Train the model
history_mnist = model_mnist.fit(x_train_mnist, y_train_mnist, batch_size=batch_size, epochs=epochs, validation_data=(x_test_mnist, y_test_mnist))

# Evaluate the model
test_loss_mnist, test_acc_mnist = model_mnist.evaluate(x_test_mnist, y_test_mnist, verbose=2)

# Plot training and validation accuracy
plt.plot(history_mnist.history['accuracy'], label='accuracy')
plt.plot(history_mnist.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

# Predictions with plots
predictions_mnist = model_mnist.predict(x_test_mnist)
for i in range(10):
    plt.imshow(x_test_mnist[i].reshape(28, 28), cmap='gray')
    plt.title(f'Actual: {y_test_mnist[i]}, Predicted: {predictions_mnist[i].argmax()}')
    plt.show()
```

With CIFAR 10

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Load and preprocess the CIFAR-10 dataset
cifar10 = tf.keras.datasets.cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Flatten the labels
y_train = y_train.flatten()
y_test = y_test.flatten()

# Normalize the images
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# Define kernel and pool sizes
kernel_size = (3, 3)
pool_size = (2, 2)

batch_size = 32
num_classes = 10
epochs = 50

classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Build the CNN model
model = models.Sequential()

# First convolutional layer
model.add(layers.Conv2D(32, kernel_size, padding='same', input_shape=x_train.shape[1:], activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Conv2D(32, kernel_size, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D(pool_size=pool_size))
model.add(layers.Dropout(0.25))

# Second convolutional layer
model.add(layers.Conv2D(64, kernel_size, padding='same', activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Conv2D(64, kernel_size, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D(pool_size=pool_size))
model.add(layers.Dropout(0.25))

# Flatten the output and add a fully connected layer
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.5))

# Output layer with softmax activation
model.add(layers.Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0001, decay=1e-06),
              loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

# Train the model
history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(x_test, y_test))

# Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

# predictions with plots
predictions = model.predict(x_test)
for i in range(10):
    plt.imshow(x_test[i])
    plt.title(f'Actual: {classes[y_test[i]]}, Predicted: {classes[predictions[i].argmax()]}')
    plt.show()
```

With CIFAR 100

```python
# Load and preprocess the CIFAR-100 dataset
cifar100 = tf.keras.datasets.cifar100
(x_train_100, y_train_100), (x_test_100, y_test_100) = cifar100.load_data()

# Flatten the labels
y_train_100 = y_train_100.flatten()
y_test_100 = y_test_100.flatten()

# Normalize the images
x_train_100 = x_train_100.astype('float32') / 255
x_test_100 = x_test_100.astype('float32') / 255

# Define kernel and pool sizes
kernel_size = (3, 3)
pool_size = (2, 2)

batch_size = 32
num_classes = 100
epochs = 100

# Build the CNN model
model_100 = models.Sequential()

# First convolutional layer
model_100.add(layers.Conv2D(32, kernel_size, padding='same', input_shape=x_train_100.shape[1:], activation='relu'))
model_100.add(layers.BatchNormalization())
model_100.add(layers.Conv2D(32, kernel_size, activation='relu'))
model_100.add(layers.BatchNormalization())
model_100.add(layers.MaxPooling2D(pool_size=pool_size))
model_100.add(layers.Dropout(0.25))

# Second convolutional layer
model_100.add(layers.Conv2D(64, kernel_size, padding='same', activation='relu'))
model_100.add(layers.BatchNormalization())
model_100.add(layers.Conv2D(64, kernel_size, activation='relu'))
model_100.add(layers.BatchNormalization())
model_100.add(layers.MaxPooling2D(pool_size=pool_size))
model_100.add(layers.Dropout(0.25))

# Flatten the output and add a fully connected layer
model_100.add(layers.Flatten())
model_100.add(layers.Dense(512, activation='relu'))
model_100.add(layers.BatchNormalization())
model_100.add(layers.Dropout(0.5))

# Output layer with softmax activation
model_100.add(layers.Dense(num_classes, activation='softmax'))

# Compile the model
model_100.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0001, decay=1e-06),
                  loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print the model summary
model_100.summary()

# Train the model
history_100 = model_100.fit(x_train_100, y_train_100, batch_size=batch_size, epochs=epochs, validation_data=(x_test_100, y_test_100))

# Evaluate the model
test_loss_100, test_acc_100 = model_100.evaluate(x_test_100, y_test_100, verbose=2)

# Plot training and validation accuracy
plt.plot(history_100.history['accuracy'], label='accuracy')
plt.plot(history_100.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

# Predictions with plots
predictions_100 = model_100.predict(x_test_100)
for i in range(10):
    plt.imshow(x_test_100[i])
    plt.title(f'Actual: {y_test_100[i]}, Predicted: {predictions_100[i].argmax()}')
    plt.show()
```
